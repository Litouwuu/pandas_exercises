### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url,sep='|')
```

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# gender_table = users.groupby(by=['occupation','gender']).user_id.count()
# type(gender_table)
# gender_table.index
# for x,y in gender_table.index:
#     print(x)
#     print(y)
#     print(gender_table[x,y])

# seperately count
male_count = users[users.gender =='M'].groupby('occupation').user_id.count()
female_count = users[users.gender == 'F'].groupby('occupation').user_id.count()
gender_df = pd.concat([male_count,female_count],axis=1,sort = True)
gender_df.columns = ['Male','Female']
gender_df.fillna(0,inplace=True)
```

```python
gender_df.Female = gender_df.Female.apply(lambda x: int(x))
gender_df['Total'] = gender_df.Male + gender_df.Female
gender_df['Male Ratio'] = gender_df.Male/gender_df.Total
gender_df.sort_values('Male Ratio',ascending=False)
```

```python
def gender_judge(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
        
users['male_number'] = users.gender.apply(gender_judge)
users.head()
a = users.groupby('occupation').male_number.sum()/users.groupby('occupation').user_id.count()
a.sort_values(ascending=False)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).age.mean().unstack()
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_info = users.groupby(['occupation','gender']).user_id.count().unstack()
gender_info = gender_info.fillna(0)
gender_info['Female ratio'] = gender_info.F / (gender_info.F+gender_info.M)
gender_info['Male ratio'] = gender_info.M / (gender_info.F+gender_info.M)
```
